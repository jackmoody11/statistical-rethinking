```python
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt


p_range = np.linspace(0, 1, 1000)
prior = np.ones(1000)
prob_data = binom.pmf(8, 15, p_range)
posterior = prob_data * prior
print("Mean: {mean}, 99%: {pct}".format(mean=(posterior * p_range).sum() / posterior.sum(), pct=np.percentile(posterior, 99)))
plt.plot(p_range, posterior)
```
